```python
import sqlite3

db_name = "/data/language_models/test_outputs/01_chunking_results/chunked_sequences.sqlite"
conn = sqlite3.connect(db_name)
sql_query = conn.execute("SELECT name from sqlite_master;")
table_names = sql_query.fetchall()
print(table_names)

for table in table_names:
    t_name = table[0]
    res = conn.execute(f"SELECT * FROM {t_name}")
    rows = res.fetchall()
    for row in rows:
        print(row[:100])
conn.close()
```
